```python
# Explore Altair package and hopefully produce some interactive charts to play with.


import pandas as pd
import numpy as np
```

```python
import altair as alt
```

```python
# Load Kenya MTF Data 

chunksize = 10

mtf_list = []
chunksize = 10
for chunk in pd.read_csv("kenya_mtf2_mini.csv", encoding='latin-1', chunksize = chunksize):
    mtf_list.append(chunk)
    
kenya_mtf2_mini = pd.concat(mtf_list, axis=0)
kenya_mtf2_mini = pd.DataFrame(data = kenya_mtf2_mini)
```

```python
kenya_mtf2_mini[['c_c_30','c_c_119']]
```

```python
# [IN PROGRESS] Primary Source Info: 
# c_c_30: What is the total duration of all the outages/blackouts in a (typical: b) week?
# c_c_119: What is your main back-up source of lighting during outages/blackouts of the grid? (Option: generator)

click = alt.selection_multi(encodings = ['color'])

primary = alt.Chart(kenya_mtf2_mini[kenya_mtf2_mini[['c_c_119', 'c_c_30']].notnull().all(1)]).mark_bar(color='green').encode(
    x = alt.X('c_c_119', 
              sort=alt.EncodingSortField(field="c_c_119", op="count", order='descending'),
              axis = alt.Axis(title='C.119')),
    y = 'count(c_c_119)',
    color = 'c_c_30',
    tooltip = 'count(c_c_30)'
).properties(width = 400, height = 400, selection = click).transform_filter(click).interactive()

primary
```
